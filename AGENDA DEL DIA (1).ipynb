Una vez que ejecutes este código, la base de datos especificada será eliminada. Si intentas ejecutarlo nuevamente, te mostrará un mensaje indicando que la base de datos ya fue eliminada anteriormente.


````python
#eliminar base de datos
from azure.cosmos import CosmosClient, exceptions

# Crea un cliente de Cosmos DB
endpoint = "https://andresdia02.documents.azure.com:443/"
key = 'ovpWKq3QnVf76g2KKnwQb2FArJz9X6lroZ3bqhvqXIkdE3rgxObPWRVWxNU6yWEwDxER5Imd983HACDbCIbnUQ=='
client = CosmosClient(endpoint, key)

# Define el ID de la base de datos que deseas borrar
database_id = 'base1'

try:
    # Obtiene una referencia a la base de datos
    database = client.get_database_client(database_id)

    # Elimina todos los contenedores dentro de la base de datos
    containers = database.list_containers()
    for container in containers:
        database.delete_container(container['id'])

    # Borra la base de datos
    client.delete_database(database)

    print(f"La base de datos '{database_id}' y todos sus contenidos han sido eliminados.")

except exceptions.CosmosResourceNotFoundError:
    # La base de datos no se encontró, por lo tanto, ya ha sido eliminada anteriormente
    print(f"La base de datos '{database_id}' no se encontró. Probablemente ya ha sido eliminada.")
except Exception as e:
    # Manejo genérico de excepciones
    print(f"Ocurrió un error al intentar eliminar la base de datos: {e}")
````

### Consulta de Posibles APIs: Costo y Beneficios

En el proceso de selección de una API, es importante considerar tanto los costos como los beneficios asociados. Aquí hay algunos puntos a tener en cuenta:

- **Costo**:
  - Evaluar el modelo de precios de la API: ¿Es de pago, freemium o completamente gratuito?
  - Determinar si hay costos ocultos, como tarifas por uso excesivo o tarifas de soporte.
  - Analizar si el costo se ajusta al presupuesto disponible y si proporciona un retorno de inversión (ROI) satisfactorio.

- **Beneficios**:
  - Investigar la funcionalidad ofrecida por la API y si satisface las necesidades del proyecto.
  - Evaluar la calidad de la documentación y el soporte proporcionado por el proveedor de la API.
  - Considerar la escalabilidad y el rendimiento de la API para garantizar su viabilidad a largo plazo.
  - Analizar las características adicionales, como la seguridad, la facilidad de integración y la comunidad de usuarios.

Al equilibrar cuidadosamente los costos y los beneficios, se puede tomar una decisión informada sobre la selección de la API que mejor se adapte a las necesidades del proyecto.


### API de Consulta de Vulnerabilidades

1. **MITRE CVE API**
   - **Descripción**: La API de MITRE CVE proporciona acceso a la Base de Datos de Vulnerabilidades y Exposiciones Comunes (CVE). Ofrece información detallada sobre vulnerabilidades conocidas, incluyendo descripciones, referencias, y enlaces a información adicional.
   - **Costo**: Gratuita.
   - **Beneficios**:
     - Amplia cobertura de vulnerabilidades conocidas.
     - Documentación completa y actualizada.
     - Actualizaciones regulares y mantenimiento por parte de MITRE.
     - Integración con otras fuentes de datos de seguridad.

2. **NVD API (National Vulnerability Database)**
   - **Descripción**: La API de NVD proporciona acceso a la base de datos nacional de vulnerabilidades mantenida por el Instituto Nacional de Estándares y Tecnología (NIST). Ofrece detalles sobre vulnerabilidades, incluyendo información sobre impacto, CVSS, y referencias.
   - **Costo**: Gratuita.
   - **Beneficios**:
     - Información detallada sobre vulnerabilidades conocidas.
     - Integración con el sistema CVSS para evaluar la gravedad de las vulnerabilidades.
     - Actualizaciones regulares y mantenimiento por parte del NIST.

3. **VulnDB API**
   - **Descripción**: La API de VulnDB proporciona acceso a una base de datos de vulnerabilidades gestionada por Risk Based Security. Ofrece datos detallados sobre vulnerabilidades, incluyendo información contextual, referencias y soluciones.
   - **Costo**: De pago, con opciones de suscripción.
   - **Beneficios**:
     - Amplia cobertura de vulnerabilidades, incluyendo vulnerabilidades no publicadas.
     - Información contextual detallada para ayudar en la comprensión y mitigación de las vulnerabilidades.
     - Soporte técnico y servicios adicionales disponibles para suscriptores.

Estas APIs proporcionan acceso a datos de vulnerabilidades que pueden ser útiles en la evaluación de riesgos y la protección de sistemas contra amenazas conocidas. Es importante evaluar cada una según las necesidades específicas del proyecto y los requisitos de seguridad.

### VULNABILIDADES MAS COMUNES

````python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo: Vulnerabilidades y su frecuencia
vulnerabilidades = ['SQL Injection', 'Cross-Site Scripting', 'Inyección de Comandos', 'Fugas de Información', 'Autenticación Débil']
frecuencia = [50, 30, 25, 20, 15]  # Frecuencia de cada vulnerabilidad

# Crear el gráfico de barras vertical
fig, ax = plt.subplots(figsize=(8, 6))  # Tamaño del gráfico
x_pos = np.arange(len(vulnerabilidades))  # Posiciones en el eje x

# Dibujar las barras
bars = ax.bar(x_pos, frecuencia, color='skyblue', edgecolor='black')

# Etiquetas en los ejes y título
ax.set_xlabel('Vulnerabilidades')  # Etiqueta del eje x
ax.set_ylabel('Frecuencia')  # Etiqueta del eje y
ax.set_title('Vulnerabilidades más Comunes')  # Título del gráfico

# Añadir valores en las barras
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

# Personalizar el eje x
ax.set_xticks(x_pos)
ax.set_xticklabels(vulnerabilidades, rotation=45, ha='right')

# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()
plt.show()
````

## Vulnerabilidades más Comunes


- **Comparación Clara**: La disposición vertical facilita la comparación entre las frecuencias de diferentes vulnerabilidades, permitiendo a los usuarios identificar rápidamente las más y menos frecuentes.

- **Legibilidad Mejorada**: La rotación de las etiquetas en el eje x a 45 grados mejora la legibilidad, incluso cuando hay muchas etiquetas. Esto garantiza que las vulnerabilidades estén claramente identificadas sin ocupar demasiado espacio en el gráfico.

- **Detalles Relevantes**: La adición de valores numéricos dentro de las barras proporciona información precisa sobre la frecuencia de cada vulnerabilidad, lo que ayuda a comprender mejor los datos presentados.

- **Estética Agradable**: La combinación de colores suaves y bordes nítidos crea una presentación visual atractiva que es fácil de entender y agradable a la vista.


````python
import matplotlib.pyplot as plt

# Datos de ejemplo: CVEs asociadas a las principales vulnerabilidades de software
cves = ['CVE-2023-1234', 'CVE-2023-5678', 'CVE-2023-9012', 'CVE-2023-3456', 'CVE-2023-7890']
frecuencia_cves = [20, 18, 15, 12, 10]  # Frecuencia de cada CVE

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
bars = plt.bar(cves, frecuencia_cves, color='lightgreen', edgecolor='black')  # Gráfico de barras vertical

# Etiquetas y título
plt.xlabel('CVEs')  # Etiqueta del eje x
plt.ylabel('Frecuencia')  # Etiqueta del eje y
plt.title('Principales CVEs Asociadas a Vulnerabilidades de Software')  # Título del gráfico

# Rotar etiquetas del eje x
plt.xticks(rotation=45, ha='right')  

# Añadir valores numéricos en las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

# Añadir rejilla en el eje y
plt.grid(axis='y', linestyle='--', alpha=0.7)  

# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()  
plt.show()  
````
