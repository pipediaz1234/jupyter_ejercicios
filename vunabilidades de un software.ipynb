```python
# Diccionario para almacenar las vulnerabilidades encontradas en VulDB
cve = {}

# Diccionario para almacenar detalles adicionales de las vulnerabilidades obtenidas de MITRE
mit = {}
```

```python
def buscar_software(nombre):
    """
    Realiza una búsqueda de vulnerabilidades asociadas a un software específico en VulDB.

    Args:
        nombre (str): El nombre del software para el que se busca vulnerabilidades.

    Raises:
        requests.exceptions.RequestException: Si hay un error en la solicitud.
    """
    url = "https://vuldb.com/?api"

    # Parámetros de la solicitud
    datos = {
        "apikey": '33d6f703126b8727526d02ce9d249789',
        "search": nombre
    }

    try:  
        print("Búsqueda del software")
        # Realizar solicitud POST a VulDB
        respuesta = requests.post(url, data=datos)
        # Convertir la respuesta a JSON
        objeto = respuesta.json()
        # Iterar sobre los resultados y guardar en el diccionario 'cve'
        for item in objeto['result']:
            cve[item['source']['cve']['id']] = item['vulnerability']
    except requests.exceptions.RequestException as e:
        print(f'Error en la solicitud: {e}')


def impAzure():
    """
    Realiza la búsqueda de detalles de las vulnerabilidades en la base de datos de MITRE.
    """
    cont = 0
    print("Función impAzure")
    # Iterar sobre las vulnerabilidades en 'cve'
    for item in cve:
        cont += 1
        if cont > 20:
            # Controlar límite de solicitudes por minuto
            time.sleep(50)
            cont = 0
        # Llamar a 'apiMitres' para obtener detalles de la vulnerabilidad
        apiMitres(item)
    print('Proceso terminado')


def apiMitres(cve_id):
    """
    Obtiene detalles adicionales de una vulnerabilidad específica utilizando el identificador CVE.

    Args:
        cve_id (str): El identificador CVE de la vulnerabilidad.
    """
    print('Búsqueda en MITRE')
    base_url = 'https://cveawg.mitre.org/api/cve'
    url = f'{base_url}/{cve_id}'
    print(f"La URL es: {url}")
    # Realizar solicitud GET a MITRE
    respuesta = requests.get(url)
    if respuesta.status_code == 200:
        # Convertir respuesta a JSON
        cve_data = respuesta.json()
        # Guardar detalles adicionales en el diccionario 'mit'
        if 'descriptions' in cve_data['containers']['cna']:
            mit[cve_id] = {'descripcion': cve_data['containers']['cna']['descriptions'], 'score': cve[cve_id]}
        else:
            mit[cve_id] = {'descripcion': None, 'score': cve[cve_id]}
    else:
        print('Información de la vulnerabilidad no encontrada')
```

```python
# Realiza una búsqueda de vulnerabilidades asociadas al software "Facebook" o cualquier software en VulDB
buscar_software('Facebook')

# Ejecuta la función para obtener detalles adicionales de las vulnerabilidades encontradas en MITRE
impAzure()

# Imprime el diccionario 'mit' en formato JSON con sangrado de 2 espacios para una mejor legibilidad
print(json.dumps(mit, indent=2))

# Imprime un mensaje indicando que el proceso ha terminado
print("Proceso terminado")

```

```python
import json
from windows_tools.installed_software import get_installed_software
import requests
import time

def buscar_vulnerabilidades(nombre_software):
    url = "https://vuldb.com/?api"
    datos = {
        "apikey": '33d6f703126b8727526d02ce9d249789',  # Reemplaza 'TU_API_KEY' con tu API key de VulDB
        "search": nombre_software
    }

    try:
        print(f"Búsqueda de vulnerabilidades para: {nombre_software}")
        respuesta = requests.post(url, data=datos)
        objeto = respuesta.json()
        if 'result' in objeto:
            vulnerabilidades = []
            for item in objeto['result'][:3]:  # Limitar a las primeras tres vulnerabilidades
                if 'cve' in item['source']:
                    cve_id = item['source']['cve']['id']
                    vulnerability = item['vulnerability']
                    score = item['source']['cve']['cvss']
                    description = item['source']['cve']['summary']
                    # Determinar el nivel de gravedad según el score (CVSS)
                    level = determinar_nivel_gravedad(score)
                    vulnerabilidades.append({"CVE": cve_id, "vulnerability": vulnerability, "score": score, "description": description, "level": level})
            return vulnerabilidades
        else:
            print("No se encontraron vulnerabilidades.")
            return []
    except requests.exceptions.RequestException as e:
        print(f'Error en la solicitud: {e}')
        return []

def determinar_nivel_gravedad(score):
    # Asignar el nivel de gravedad según el rango del score
    if score >= 9.0:
        return "Peligrosa"
    elif score >= 7.0:
        return "Alta"
    elif score >= 4.0:
        return "Media"
    elif score > 0.0:
        return "Baja"
    else:
        return "Desconocida"

software_details = []

for software in get_installed_software():
    vulnerabilidades = buscar_vulnerabilidades(software['name'])
    software_details.append({"name": software['name'], "version": software['version'], "publisher": software['publisher'], "vulnerabilities": vulnerabilidades})

time.sleep(2)

print("Detalles del software instalado:")
for software in software_details:
    print(json.dumps(software, indent=4))
```
